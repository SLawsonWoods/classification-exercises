Load the attendance.csv file and calculate an attendnace percentage for each student. One half day is worth 50% of a full day, and 10 tardies is equal to one absence.


```python
df.info
```

```python
#melt df
df_long =df.melt(id_vars='Unnamed: 0', var_name = 'date',
                 value_name='status')
```

```python
#rename columns
df_long.columns =['name','date','status']
```

```python
df_long
```

```python
#map, create a dictionary
attendance_policy = {
    'H': 0.5,
    'T': 0.9,
    'A': 0,
    'P': 1
}
```

```python
df_long['attendance']= df_long.status.map(attendance_policy)
```

```python
df_long.head()
```

```python
df_long.groupby('name').attendance.mean()
```
